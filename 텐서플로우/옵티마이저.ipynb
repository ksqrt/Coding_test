```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
```

```python
# 1 input -> 3 output 
# 랜덤시드 미리 결정 
tf.random.set_seed(0)

model = keras.Sequential([keras.layers.Dense(units=3, input_shape=[1])])
```

```python
# 손실함수는 mse 옵티마이저는 SGD
# SGD 는 확률적 경사 하강법 
model.compile(loss="mse",optimizer="SGD")
```

```python
# fit 함수는 훈련 진행 상황과 현재의 손실값을 반환
model.fit([1],[[0,1,0]],epochs=1)
# eval 함수는 손실값을 확인 
model.evaluate([1], [[0, 1, 0]])
```

```python
history = model.fit([1], [[0, 1, 0]], epochs=100)
```

```python
# 손실값 시각화 색칠하기
import matplotlib.pyplot as plt

plt.style.use('default')
plt.rcParams['figure.figsize'] = (4, 3)
plt.rcParams['font.size'] = 12

loss = history.history['loss']
plt.plot(loss)
plt.text(-0.4, 0.8, 'Epoch 0')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show()
```

```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import numpy as np

# 플롯의 기초 틀잡기
plt.style.use('default')
plt.rcParams['figure.figsize'] = (4, 3)
plt.rcParams['font.size'] = 12

tf.random.set_seed(0)

# 1->3 레이어 구성
model = keras.Sequential([keras.layers.Dense(units=3, input_shape=[1], use_bias=False)])
# 손실함수, 옵티마이저 설정
model.compile(loss='mse', optimizer='SGD')

pred = model.predict([1])
print(pred)
print(model.get_weights())

plt.bar(np.arange(3), pred[0])
# y값 범위 설정
plt.ylim(-1.1, 1.1)
plt.xlabel('Output Node')
plt.ylabel('Output')
# 이 위치에다 텍스트 적기
plt.text(-0.4, 0.8, 'Epoch 0')
plt.tight_layout()
plt.savefig('./plt/pred000.png')
plt.clf()

epochs = 500
for i in range(1, epochs+1):
  model.fit([1], [[0, 1, 0]], epochs=1, verbose=0)
  pred = model.predict([1])

  if i % 25 == 0:
      plt.bar(np.arange(3), pred[0])
      plt.ylim(-1.1, 1.1)
      plt.xlabel('Output Node')
      plt.ylabel('Output')
      plt.text(-0.4, 0.8, 'Epoch ' + str(i))
      plt.tight_layout()
      plt.savefig('./plt/pred' + str(i).zfill(3) + '.png')
      plt.clf()

print(pred)
print(model.get_weights())
```
