```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
```

```python
# 1 input -> 3 output 
# 랜덤시드 미리 결정 
tf.random.set_seed(0)

model = keras.Sequential([keras.layers.Dense(units=3, input_shape=[1])])
```

```python
# 손실함수는 mse 옵티마이저는 SGD
# SGD 는 확률적 경사 하강법 
model.compile(loss="mse",optimizer="SGD")
```

```python
# fit 함수는 훈련 진행 상황과 현재의 손실값을 반환
model.fit([1],[[0,1,0]],epochs=1)
# eval 함수는 손실값을 확인 
model.evaluate([1], [[0, 1, 0]])
```

```python
history = model.fit([1], [[0, 1, 0]], epochs=100)
```

```python
# 손실값 시각화 색칠하기
import matplotlib.pyplot as plt

plt.style.use('default')
plt.rcParams['figure.figsize'] = (4, 3)
plt.rcParams['font.size'] = 12

loss = history.history['loss']
plt.plot(loss)
plt.text(-0.4, 0.5, 'Epoch 0')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show()
```
