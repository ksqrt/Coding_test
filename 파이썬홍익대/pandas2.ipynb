```python
import pandas as pd

df = pd.read_csv("data/@preprocessing_data_member01.csv")

df.head(7)
df.tail(7)
```

```python
df.pivot_table(index = "유입경로",columns="성별",values="총판매액",aggfunc="sum")
```

```python
df.describe()
```

```python
df["성별(replaced)"]= df.replace(["남자","여자"],[1,0])["성별"]

df.tail()
```

```python
df.groupby("가입연도").sum()["총판매액"]
s =df.groupby(["가입연도","ID"]).sum()["총판매액"]
s = s.reset_index()

s[s["가입연도"]==2019].sort_values("총판매액",ascending=True).head()["ID"]
```

```python
pv =df.pivot_table(index= "ID",columns="가입연도",values="총판매액",aggfunc=sum)

pv.sort_values(by = 2019,ascending=True).head()
```

```python
pv =df.pivot_table(index="ID" ,columns="유입경로",values="총구매횟수",aggfunc=sum)
pv.sort(by="총구매횟수" ,ascending=True)
```

```python
s =df.groupby(["최종주문요일","유입경로"]).sum()["총판매액"]
s = s.reset_index().sort_values("총판매액",ascending=False)
s
```
